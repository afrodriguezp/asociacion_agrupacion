##Análisis de Componentes Principales (PCA):

```python
import numpy as np
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

# Seleccionar las características numéricas relevantes
features = ['weight', 'appraisal', 'payment', 'cost', 'rate']

# Crear una matriz de características
X = reread[features]

# Escalar las características (opcional pero a menudo recomendado)
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Crear un objeto PCA y ajustarlo a los datos
pca = PCA(n_components=2)  # Elige el número de componentes principales que deseas
principal_components = pca.fit_transform(X_scaled)


# Imprimir la varianza explicada por cada componente
print("Varianza explicada por cada componente:", pca.explained_variance_ratio_)


# Gráfico de dispersión de los datos reducidos
plt.scatter(principal_components[:, 0], principal_components[:, 1])
plt.xlabel("Componente Principal 1")
plt.ylabel("Componente Principal 2")
plt.title("Gráfico de Dispersión de PCA")
plt.show()

# Crear un mapa de calor
sns.heatmap(principal_components, cmap='coolwarm')
plt.title('Mapa de Calor de Componentes Principales')
plt.show()


cumulative_variance_ratio = np.cumsum(pca.explained_variance_ratio_)

# Graficar la varianza explicada acumulativa
plt.plot(range(1, len(cumulative_variance_ratio) + 1), cumulative_variance_ratio, marker='o', linestyle='--')
plt.xlabel('Número de Componentes Principales')
plt.ylabel('Varianza Explicada Acumulativa')
plt.title('Varianza Explicada Acumulativa por Componentes Principales')
plt.show()
```

Validación con indices de:

####Silhouette Score:
El Silhouette Score mide cuán similar es un objeto a su propio clúster (cohesión) en comparación con otros clústeres vecinos (separación). El valor del Silhouette Score varía de -1 a 1, donde:
1: Indica que el objeto está bien ajustado a su propio clúster y mal ajustado a clústeres vecinos.
0: Indica que el objeto está en o muy cerca del límite de decisión entre dos clústeres.
-1: Indica que el objeto puede haber sido asignado incorrectamente.
En general, un Silhouette Score más alto sugiere una mejor calidad de los clústeres.
####Davies-Bouldin Score:
El Davies-Bouldin Score mide la compacidad intraclúster y la separación interclúster. Un valor más bajo del Davies-Bouldin Score indica clústeres más compactos y mejor separados. Se calcula como el promedio de las relaciones de similitud entre cada clúster y su clúster más similar.
A diferencia del Silhouette Score, el Davies-Bouldin Score no asume una geometría esférica de los clústeres y puede ser útil cuando los clústeres tienen formas más complejas.

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn.metrics import precision_score, recall_score, f1_score

# Función para aplicar K-means y evaluar métricas

X = df[['weight', 'appraisal', 'payment','cost','rate','percentageUpfront','vehicleEncoded','originEncoded','destinationEncoded']].values


def apply_kmeans_and_evaluate(k):
    # Aplicar K-means
    kmeans = KMeans(n_clusters=k, random_state=42)
    labels = kmeans.fit_predict(X)

    # Calcular métricas internas (Coeficiente de Silhouette y Davies-Bouldin)
    silhouette_score = metrics.silhouette_score(X, labels)
    davies_bouldin_score = metrics.davies_bouldin_score(X, labels)


    return silhouette_score, davies_bouldin_score

# Probar con k=2, k=3, k=4, k=5, k=6, k=7 y k=8
k_values = [2, 3, 4, 5, 6, 7, 8]
results = []

for k in k_values:
    metrics_results = apply_kmeans_and_evaluate(k)
    results.append((k,) + metrics_results)

# Mostrar los resultados
columns = ['K', 'Silhouette Score', 'Davies-Bouldin Score']
results_df = pd.DataFrame(results, columns=columns)
print("Kmeans")
print(results_df)
```

```python
from sklearn import metrics
from sklearn.metrics import pairwise_distances
from sklearn.metrics import precision_score, recall_score, f1_score

# Función para aplicar K-means y evaluar métricas
X = df[['weight', 'appraisal', 'payment','cost','rate','percentageUpfront','vehicleEncoded','originEncoded','destinationEncoded']].values

def apply_DBSCAN_and_evaluate(k):

    # Crear un objeto de Clustering Jerárquico Divisivo
    eps = 70000000
    min_samples = int((len(X)/80)/k)
    dbscan = DBSCAN(eps=eps, min_samples=min_samples)

    # Ajustar el modelo a los datos y Etiquetas de grupo asignadas por el modelo
    labels = dbscan.fit_predict(X)

    # Para saber cuántos clústeres distintos hay (excluyendo el ruido):
    unique_labels = np.unique(labels)
    num_clusters = len(unique_labels)  # Restamos 1 para excluir el ruido (etiqueta -1)
    print("Número de clústeres:", num_clusters)

    # Calcular métricas internas (Coeficiente de Silhouette y Davies-Bouldin)
    silhouette_score = metrics.silhouette_score(X, labels)
    try:
      davies_bouldin_score = metrics.davies_bouldin_score(X, labels)
    except ValueError as e:
      print("No es posible calcular el Davies-Bouldin Score para DBSCAN:", str(k))

    return silhouette_score, davies_bouldin_score

# Probar con k=2, k=3, k=4, k=5, k=6, k=7 y k=8
k_values = [2, 3, 4, 5, 6, 7, 8]
results = []

for k in k_values:
    metrics_results = apply_DBSCAN_and_evaluate(k)
    results.append((k,) + metrics_results)

# Mostrar los resultados
columns = ['K', 'Silhouette Score', 'Davies-Bouldin Score']
results_df = pd.DataFrame(results, columns=columns)
print("DBSCAN")
print(results_df)
```

Para el algoritmo DBSCAN (Density-Based Spatial Clustering of Applications with Noise) es un poco diferente en comparación con los algoritmos de agrupamiento como K-means. Esto se debe a que DBSCAN no asigna todos los puntos a un clúster y puede tener puntos clasificados como ruido.

```python
from efficient_apriori import apriori

data = reread
# Transformación de datos en transacciones
transactions = []
for _, row in data.iterrows():
    transaction = [str(val) for val in row if not pd.isna(val)]
    transactions.append(transaction)

# Ejecución del algoritmo Apriori
itemsets, rules = apriori(transactions, min_support=0.2, min_confidence=0.7)

# Aplicar Apriori para encontrar conjuntos frecuentes con un soporte mínimo de 2
#print(3/len(transactions))
#itemsets, rules = apriori(transactions, min_support=3/len(transactions))
print(itemsets)
# Mostrar los conjuntos frecuentes
print("Conjuntos frecuentes:")
for itemset in itemsets:
    print(itemset)

# Mostrar las reglas de asociación
# Imprime las reglas de asociación
print("Número de reglas" + str(len(rules)))
for rule in rules:
    print(f"Regla: {rule.lhs} -> {rule.rhs}")
    print(f"Soporte: {rule.support}")
    print(f"Confianza: {rule.confidence}")
    print("\n")
```
